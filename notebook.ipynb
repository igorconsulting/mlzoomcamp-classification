### Make `median_house_value` binary

* We need to turn the `median_house_value` variable from numeric into binary.
* Let's create a variable `above_average` which is `1` if the `median_house_value` is above its mean value and `0` otherwise.

```python
Mean = data['median_house_value'].mean()

data['above_average'] = 0

data.loc[data['median_house_value']> Mean, 'above_average'] = 1
```

### Split the data

* Split your data in train/val/test sets, with 60%/20%/20% distribution.
* Use Scikit-Learn for that (the `train_test_split` function) and set the seed to 42.
* Make sure that the target value (`median_house_value`) is not in your dataframe.

```python
from sklearn.model_selection import train_test_split
```

```python
# Let's say we want to split the data in 80:10:10 for train:valid:test dataset
import random as r
Seed = r.seed(42)
train_size=0.6

X = data.drop(columns = ['above_average']).copy()
y = data['above_average']

# In the first step we will split the data in training and remaining dataset
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.6,random_state = Seed)

# Now since we want the valid and test size to be equal (20% each of overall data). 
# we have to define valid_size=0.5 (that is 50% of remaining data)
test_size = 0.5
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5,random_state = Seed)

print(X_train.shape), print(y_train.shape)
print(X_valid.shape), print(y_valid.shape)
print(X_test.shape), print(y_test.shape)
```

```python
import numpy as np
def data_transform(datas: pd.core.frame.DataFrame) -> pd.core.frame.DataFrame:
    ohe = OneHotEncoder()
    A = np.asarray(datas['ocean_proximity']).reshape(-1,1)
    feature_array = ohe.fit_transform(A).toarray()
    feature_label = ohe.categories_
    feature_label = np.unique(A).ravel()

    # create DataFrame
    B = datas
    B[feature_label] = pd.DataFrame(feature_array)

    B = B.drop(['ocean_proximity'], axis = 1)
    return B
```

```python

ohe = OneHotEncoder()
A = np.asarray(data['ocean_proximity']).reshape(-1,1)
feature_array = ohe.fit_transform(A).toarray()
feature_label = ohe.categories_
feature_label = np.unique(A).ravel()

# create DataFrame
```

```python
import random as r
Seed = r.seed(42)
train_size=0.6
data = data.dropna()
X = data.drop(columns = ['above_average']).copy()
y = data['above_average']

# In the first step we will split the data in training and remaining dataset
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.6,random_state = Seed)

# Now since we want the valid and test size to be equal (20% each of overall data). 
# we have to define valid_size=0.5 (that is 50% of remaining data)
test_size = 0.5
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5,random_state = Seed)

print(X_train.shape), print(y_train.shape)
print(X_valid.shape), print(y_valid.shape)
print(X_test.shape), print(y_test.shape)
```
